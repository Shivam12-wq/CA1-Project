```python
import pandas as pd
import plotly.express as px

# Filter data excluding "Unknown" continent
filtered_data = data[data['continent'] != 'Unknown']

# Group data by continent and location, then sort within each group by total_cases
top_countries_by_continent = filtered_data.groupby(['continent', 'location']).agg(
    total_cases=('total_cases', 'max'),
    population=('population', 'max'),
    total_deaths=('total_deaths', 'max')
).reset_index()

# Get top 5 COVID-19 affected countries from each continent based on the number of cases
top_countries_by_continent = top_countries_by_continent.groupby('continent').apply(lambda x: x.nlargest(5, 'total_cases'))

# Plotting the comparison of population and total deaths for top affected countries from each continent
# Create a bubble chart using Plotly Express
fig = px.scatter(top_countries_by_continent, x='population', y='total_deaths', size='total_cases', color='continent',
                 hover_name='location', size_max=50, labels={'population': 'Total Population', 'total_deaths': 'Total Deaths'})

# Update layout and display the figure
fig.update_layout(title='Comparison of Population vs Total Deaths for Top 5 Affected Countries per Continent',
                  xaxis_title='Total Population', yaxis_title='Total Deaths (COVID-19)')
fig.show()

```

**(f): Apply dummy encoding to categorical variables (at least one variable used from the data set) and
 discuss the benefits of dummy encoding to understand the categorical data.**<br><br>



Dummy encoding is a technique used to convert categorical variables into a numerical format suitable for machine learning algorithms. It involves creating binary (0 or 1) columns for each category present in the categorical variable. This transformation allows algorithms to interpret and process categorical data effectively.
 <br><br>

**Benefits of Dummy Encoding:**<br><br>

**1. Compatibility with Algorithms:** Many machine learning algorithms require numerical input. Dummy encoding converts categorical data into a format that algorithms can interpret, allowing their application to a broader range of models.<br><br>

**2. Preserves Information:** Dummy encoding represents categorical variables without assigning any ordinal or hierarchical structure. Each category becomes an independent binary feature, preserving the individuality of categories.
<br><br>
**3. No Assumptions on Relationship:** Unlike label encoding, which assigns ordinal values to categories, dummy encoding does not imply any order or relationship between categories, avoiding potential misinterpretation by algorithms.
<br><br>
**4. Prevents Magnitude Impact:** Dummy encoding ensures that no false ordinal relationships or magnitudes are assigned to categories, thereby preventing the model from making incorrect assumptions based on category labels.
<br><br>
**5. Improved Model Performance:** By converting categorical data into a numerical format that algorithms can understand, dummy encoding enhances the predictive power of machine learning models.

```python
# Let's apply dummy encoding to a categorical variable from the our dataset.
# We have a categorical variable 'continent' in the dataset 'data'.
# We are applying dummy encoding using Pandas:

# Apply dummy encoding using Pandas
dummy_encoded_data = pd.get_dummies(data['continent'], prefix='continent')

# Concatenate the dummy encoded columns with the original data
data_with_dummies = pd.concat([data, dummy_encoded_data], axis=1)

# Drop the original 'continent' column as it's no longer needed
data_with_dummies.drop('continent', axis=1, inplace=True)

# Display the resulting dataset with dummy encoding
print(data_with_dummies.head())
```
